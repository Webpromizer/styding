```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-10, 10)
y = x**2
fix, ax = plt.subplots()
ax.plot(x,y)
```

```python
mu, sigma = 1, 4  
s = np.random.normal(mu, sigma, 1000)  

plt.hist(s, bins=30, density=True, alpha=0.7, color='g')  # bins - количество бинов, density - нормализация
plt.xlabel('Y-координата')
plt.ylabel('X-координата')
plt.title('Гистограмма нормального распределения')
plt.grid(True)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(5,5))
x = np.arange(0, 10, 0.1)

axs[0, 0].plot(x, x) 
axs[0, 1].plot(x, x**2)
axs[1, 0].plot(x, x**3) 
axs[1, 1].plot(x, np.sqrt(x))
```

```python
fig, ax = plt.subplots(1, 3, figsize=(8,3))
s_1 = [23, 12, 56]
s_2 = [56, 43, 23]
s_3 = [10, 53, 33]
ax[0].bar(np.arange(len(s_1)), s_1, color="g", ec="b", lw=2, label="начальник")
ax[0].legend(loc='upper left', bbox_to_anchor=(0, 1))
ax[1].bar(np.arange(len(s_2)), s_2, color="r", label="сотрудник")
ax[1].legend(loc='upper right', bbox_to_anchor=(1, 1))
ax[2].bar(np.arange(len(s_3)), s_3, color="y", label="менеджер")
ax[2].legend(loc='upper right', bbox_to_anchor=(1, 1))
```
